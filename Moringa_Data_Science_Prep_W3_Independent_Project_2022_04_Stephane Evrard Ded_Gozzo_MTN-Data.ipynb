IMPORTING PYTHON LIBRARIES

```python
# Let's import the pandas library

import pandas as pd

# Let's also import the Numpy library
import numpy as np
```

LOADING DATASETS

```python
# LOADING TELECOM DATASET 1

#Telecom dataset 1

df1 = pd.read_csv('Telcom_dataset.csv')

df1.head(5)

```

```python
# LOADING TELECOM DATASET 2

#Telecom dataset 2

df2 = pd.read_csv('Telcom_dataset2.csv')

df2.head(5)
```

```python
# LOADING TELECOM DATASET 3

#Telecom dataset 3

df3 = pd.read_csv('Telcom_dataset3.csv')

df3.head(5)
```

LOADING OF CELLS_GEO DATASET

```python
# Loading our cells_geo

df5 = pd.read_csv('cells_geo.csv' ,index_col=0 , delimiter = ';')

df5.head(5)
```

LET'S CLEAN OUR DATA

```python
# We have to rename our Dataset1 and Dataset3 to fit with the Dataset2

# Renaming of Dataset1

df1 = pd.read_csv('Telcom_dataset.csv')
df1.rename(columns={'PRODUTC' :'PRODUCT' , 'DW_A_NUMBER_INT': 'DW_A_NUMBER', 'DW_B_NUMBER_INT': 'DW_B_NUMBER','DATETIME' : 'DATE_TIME'}, inplace=True)


df1.head(5)
```

```python
# Renaming of Dataset3

df3 = pd.read_csv('Telcom_dataset3.csv')
df3.rename(columns={'DW_A_NUMBER_INT': 'DW_A_NUMBER', 'DW_B_NUMBER_INT': 'DW_B_NUMBER','CELLID' : 'CELL_ID' , 'SIET_ID' : 'SITE_ID'}, inplace=True)

df3.head(5)
```

CONCATENATING OF THREE DATASETS

```python
# combining the three datasets using concatenation

df4 = pd.concat([df1 ,df2,df3])

df4
```

DROPPING USELESS COLUMNS

```python
# Dropping unsuable columns 

df4 = df4.drop(['COUNTRY_A',"COUNTRY_B",'CELL_ON_SITE'] , axis = 1)

df4.head()
```

MERGING OUR TELECOM_DATASET 1, 2, AND 3 WITH CELLS_GEO

```python
# Merging our Telecom dataset with cell_geo

# df4 is a concatenation of df1, df2, and df3
# df5 is for cell_geo

df_all = df4.merge(df5 , left_index=True , right_index=True)
df_all.head()
```

```python
# Let's separate our date_time in date and time
# For that we use no name function (lambda)

df1 = df_all['DATE_TIME'].apply(lambda y: pd.Series([i for i in reversed(y.lower().split(' '))]))
df1.columns = ['TIME', 'DATE']
df1
df6 = df1.merge(df_all, left_index=True, right_index=True)
df6

#droping DATE_TIME column
df7 = df6.drop('DATE_TIME', axis=1)
df7.head()

df7.head()
```

REMOVING THE NULL VALUE IN ROWS

```python
# removing the rows that contains NULL values

df_all = df_all.dropna()
df_all.head(5)
```

REMOVING THE THE DUPLICATE DATA IN OUR DATASETS

```python
# Let's take off identical data

df_all = df_all.drop_duplicates()
df_all.head()
```

City where they spend more during business hours

```python
# Which cities  used most expenditure  during business  hours?

df_time= df_all[(df_all['DATE_TIME'] > '08.00.00' ) & (df_all['DATE_TIME'] < '17.00.00')]
list(set(df_time['VILLES']))

group_city = df_time.groupby(['VILLES', 'PRODUCT'])['VALUE'].max()

group_city
```

City where they spend more during home hours


```python
# Which cities they spend more during  home hours?

df_time1= df_all[ (df_all['DATE_TIME'] < '05.00.00') ]
df_time2 = df_all[(df_all['DATE_TIME'] > '17.00.00') ]

both = pd.concat([df_time1, df_time2])
group = both.groupby(['VILLES', 'PRODUCT'])['VALUE'].max()

group
```
